```python
import math

class Figura:
    
    def area(self):
        raise NotImplementedError("Método abstrato que deve ser implementado por uma classe filha")
        
class Circulo(Figura):
    
    def __init__(self, raio):
        self.raio = raio
    
    def area(self):
        return math.pi * (self.raio**2)
    
class Poligono(Figura):
    
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura        
    
        
class Triangulo(Poligono):
    
    def area(self):
        return (self.base * self.altura) / 2

class Losango(Poligono):
    
    def area(self):
        return (self.base * self.altura) / 2
    
class Diagonal:
    
    def diagonal(self, base, altura):
        return (base**2 + altura**2)**(1/2)
    

class Retangulo(Poligono):
    
    diagonal = Diagonal()
        
    def diagonal(self):
        return self.diagonal.diagonal(self.base, self.altura)
    
    def area(self):
        return self.base * self.altura

class Quadrado(Retangulo):
    def __init__(self, lado):
        Retangulo.__init__(self, lado, lado)

raioCirculo = float(input('Coloque o raio do circulo:'))
baseTriangulo = float(input('Coloque a base do triangulo:'))
alturaTriangulo = float(input('Coloque a altura do triangulo:'))
baseLosango = float(input('Coloque a base do losango:'))
alturaLosango = float(input('Coloque a altura do losango:'))
baseRetangulo = float(input('Coloque a base do retangulo:'))
alturaRetangulo = float(input('Coloque a altura do retangulo:'))
ladoQuadrado = float(input('Coloque o lado do quadrado:'))

        
class Geometria:
    
    def __init__(self, figuras):
        self.lista = [x for x in figuras]
        
geometria = Geometria([Circulo(raioCirculo),
                       Triangulo(baseTriangulo, alturaTriangulo),
                       Losango(baseLosango, alturaLosango),
                       Retangulo(baseRetangulo,alturaRetangulo),
    Quadrado(ladoQuadrado)
])


for x in geometria.lista:
    print(f'Figura: {type(x).__name__} \nArea: {x.area()}\n')
```
